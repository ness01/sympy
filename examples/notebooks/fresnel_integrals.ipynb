```python
from numpy import linspace
from sympy import Symbol, fresnel_S, fresnel_C, oo
from matplotlib.pyplot import figure, plot, xlim, ylim
```

```python
t = linspace(0, 8, 8000)
```

```python
u = [ fresnel_S(i).evalf() for i in t ]
```

```python
v = [ fresnel_C(i).evalf() for i in t ]
```

```python
u = array(u)
v = array(v)
```

Plot of the two Fresnel integrals $S(x)$ and $C(x)$

```python
figure(figsize=(10,8))
plot(t, u)
plot(t, v)
```

The Cornu spiral defined as the parametric curve $u(t),v(t) := C(t), S(t)$

```python
figure(figsize=(10,8))
plot(v, u, "b")
plot(-v, -u, "b")
grid(True)
```

```python
z = Symbol("z")
```

Compute and plot the leading order behaviour around $x=0$

```python
ltc = fresnel_C(z).as_leading_term(z)
```

```python
lts = fresnel_S(z).as_leading_term(z)
```

```python
lts, ltc
```

```python
lt = linspace(0, 1.5, 1000)
```

```python
ltu = [ lts.subs(z, i).evalf() for i in lt ]
```

```python
ltv = [ ltc.subs(z, i).evalf() for i in lt ]
```

```python
figure(figsize=(10,8))
plot(t, u, "b")
plot(t, v, "b")
plot(lt, ltu, "r")
plot(lt, ltv, "r")
xlim(0, 1)
ylim(0, 0.8)
```

Compute and plot the asymptotic series expansion at $x=\infty$

```python
ats = fresnel_S(z)._eval_aseries(z, oo, 0,0)
```

```python
atc = fresnel_C(z)._eval_aseries(z, oo, 0,0)
```

```python
ats, atc
```

```python
at = linspace(6, 8, 2000)
```

```python
ass = [ ats.subs(z, i).evalf() for i in at ]
```

```python
asc = [ atc.subs(z, i).evalf() for i in at ]
```

```python
figure(figsize=(10,8))
plot(t, u, "b")
plot(t, v, "b")
plot(at, ass, "r")
plot(at, asc, "r")
xlim(6,8)
ylim(0.4, 0.6)
```
